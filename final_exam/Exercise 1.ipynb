```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('redwinedata/redwine_train.txt')
labels = np.loadtxt('redwinedata/redwine_trainlabels.txt')
good = []
bad = []
for i in range(len(labels)):
    if labels[i] == 1.0:
        good.append(data[i,10])
    else:
        bad.append(data[i,10])
        
wines = []
wines.append(good)
wines.append(bad)

plt.boxplot(wines, labels=["Good","Bad"])
plt.title('Boxplot over alcohol levels for good and bad wines')
plt.ylabel('Alcohol levels')
```

```python
plt.hist(good, 20, label='Good')
plt.hist(bad, 20, color='red', label='Bad')
plt.xlabel('Alcohol levels')
plt.ylabel('Count')
plt.title('Histogram over alcohol levels in good and bad wines')
plt.legend()
```
